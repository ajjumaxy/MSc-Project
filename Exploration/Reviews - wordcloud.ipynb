## Settings

```python
reviews_path = 'C:\\Data\\UCL\\@MSc Project - Data and sources\\reviews.csv'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from os.path import join
from datetime import datetime

from wordcloud import WordCloud

%matplotlib inline
```

## Load the data

```python
reviews = pd.read_csv(reviews_path)

min_date = datetime.strptime('2018-7-1', '%Y-%m-%d')
max_date = datetime.strptime('2020-6-30', '%Y-%m-%d')
reviews['Date'] = pd.to_datetime(reviews.Date)

reviews = reviews[(reviews.Date >= min_date) & (reviews.Date <= max_date)]

reviews['Review'] = reviews['Pros'] + ' ' + reviews['Cons']
reviews['Review'] = reviews['Review'].astype('str')
```

## WordCloud

```python
text = ' '.join(reviews['Review'])

wordcloud = WordCloud(
    width=1000,
    height=1000,
    background_color='white',
    min_font_size = 15,
    max_font_size = 100,
    max_words = 100,
).generate(text)
```

```python
plt.figure(figsize=(12,12))

plt.imshow(wordcloud, facecolor=None)
plt.axis("off")
plt.show()
```
